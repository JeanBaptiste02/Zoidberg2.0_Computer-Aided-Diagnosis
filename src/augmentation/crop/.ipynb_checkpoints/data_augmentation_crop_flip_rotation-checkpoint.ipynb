## Objectif

Ce notebook a pour objectif d'explorer l'effet de la technique de Data Augmentation par recadrage (Crop), retournement horizontal (Flip) et rotation dans l'entraînement d'un modèle de réseau de neurones convolutifs (CNN) pour la classification d'images. Ces techniques permettent d'augmenter la diversité des données d'entraînement, ce qui améliore la robustesse et la généralisation du modèle. L'ensemble de données utilisé pour cette étude est un dataset de rayons X pour la détection de pathologies pulmonaires, structuré en trois ensembles : entraînement, validation, et test.

## Étapes de Démarche

1. **Vérification de la Disponibilité du GPU**
   - Validation de la présence d'un GPU pour accélérer l'entraînement du modèle.
2. **Chargement et Préparation des Données**
   - Chargement du dataset CIFAR-10, redimensionnement des images, conversion en niveaux de gris, normalisation des valeurs des pixels, et restructuration des images pour les adapter à l'entrée du modèle.
3. **Visualisation des Données**
   - Aperçu visuel des images avant augmentation, permettant de comprendre les données de base.
4. **Augmentation des Données**
   - Application de la technique de recadrage central, en combinaison avec d'autres transformations comme le zoom, la translation et la rotation, pour augmenter la variabilité des données d'entraînement.
5. **Définition et Entraînement du Modèle CNN**
   - Construction d'un modèle CNN comprenant des couches de convolution, de pooling, et des couches fully connected. Entraînement du modèle avec les données augmentées.
6. **Évaluation du Modèle**
   - Test du modèle sur les données de validation pour évaluer sa précision et sa capacité de généralisation.
7. **Visualisation des Performances**
   - Analyse approfondie des performances du modèle à l'aide de différents outils :
     - **Matrice de confusion** : Visualisation des erreurs de classification.
     - **Rapport de classification** : Evaluation des performances pour chaque classe.
     - **Courbe ROC** : Analyse de la capacité du modèle à distinguer les différentes classes à divers seuils.
8. **Prédictions et Analyse des Erreurs**
     - Les prédictions sur les données d'entraînement ont été affichées pour visualiser les résultats. Les images avec leurs étiquettes prédites et
       réelles ont été montrées pour analyser les erreurs.
     - Histogramme des Prédictions : Un histogramme des probabilités prédites a été tracé pour comprendre la distribution des prédictions du modèle.
     - Tableau des Prédictions : Un tableau montrant les prédictions, les probabilités et les vraies étiquettes a été imprimé pour une inspection détaillée.

### Configuration des Paramètres d'Entraînement et des Chemins des Données

Ce code initialise les paramètres de configuration pour le traitement des images et l'entraînement du modèle, notamment la taille des images, les chemins vers les répertoires de données, ainsi que la taille des lots et le nombre d'époques d'entraînement.

```python
img_size = 150
data_dir1 = '../../../chest_xray/train/PNEUMONIA'
data_dir2 = '../../../chest_xray/train/NORMAL'
print("folders")
batch_size = 32
epochs = 20
```

### Chargement des Données

Le code suivant définit une fonction pour charger les données d'images à partir des répertoires spécifiés, redimensionner les images et les préparer pour l'entraînement. La fonction traite les images en niveaux de gris et les étiquette en fonction de leur répertoire.

- **Fonction `load_data`** :
  - **Paramètres** :
    - `data_dir1` : Répertoire contenant les images de la première classe.
    - `data_dir2` : Répertoire contenant les images de la deuxième classe.
    - `limit` : Nombre maximum d'images à charger par répertoire (fixé à 100 dans cet exemple).
  - **Étapes** :
    - **Chargement des données** :
      - Vérifie si les répertoires existent.
      - Associe chaque image à un label basé sur son répertoire.
      - Lit chaque image en niveaux de gris, la redimensionne, et l'ajoute à la liste des données.
    - **Gestion des erreurs** :
      - Informe si un répertoire n'existe pas ou si une image échoue à être lue.
  
- **Chargement des données d'entraînement** :
  - La fonction est appelée pour charger les données depuis les répertoires spécifiés avec une limite de 100 images par répertoire.


```python
def load_data(data_dir1, data_dir2, limit=100):
    print("Loading data...")
    data = []
    labels = ['PNEUMONIA', 'NORMAL']
    for dir in [data_dir1, data_dir2]:
        path = os.path.join(dir)
        if not os.path.exists(path):
            print(f"Le répertoire {path} n'existe pas")
            continue
        label = os.path.basename(dir)
        class_num = labels.index(label)
        for i, img in enumerate(os.listdir(path)):
            if i >= limit:
                break
            try:
                img_arr = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
                if img_arr is None:
                    print(f"Failed to read {img}. Skipping.")
                    continue
                resized_arr = cv2.resize(img_arr, (img_size, img_size))
                data.append([resized_arr, class_num])
            except Exception as e:
                print(e)
    print("Finished loading data.")
    return np.array(data, dtype=object)

# Chargement des données
train_data = load_data(data_dir1, data_dir2, limit=100)
```

### Prétraitement des Données

Le code suivant définit des fonctions pour prétraiter les images et séparer les caractéristiques des étiquettes, puis applique ces étapes aux données d'entraînement.

- **Fonctions** :

  - **`separate_features_labels(data)`** :
    - Sépare les caractéristiques (images) et les étiquettes à partir des données chargées.
    - **Paramètre** : `data` - Données sous forme de liste d'images et d'étiquettes.
    - **Retour** : Deux tableaux NumPy : un pour les caractéristiques (images) et un pour les étiquettes.

  - **`preprocess_image(image)`** :
    - Prétraite une image individuelle en l'ajustant à la taille spécifiée et en normalisant les valeurs de pixel.
    - **Paramètre** : `image` - Image à prétraiter.
    - **Retour** : Image redimensionnée et normalisée.

  - **`preprocess_data(data)`** :
    - Applique le prétraitement à l'ensemble des données.
    - **Paramètre** : `data` - Données contenant des images et des étiquettes.
    - **Retour** : Données prétraitées, avec des images normalisées et redimensionnées.

- **Application des Fonctions** :
  - **Chargement des données** :
    - `train_data = load_data(data_dir1, data_dir2, limit=100)` : Chargement des données d'entraînement.
  - **Prétraitement des données** :
    - `train_data = preprocess_data(train_data)` : Prétraitement des images et étiquettes.
  - **Séparation des caractéristiques et des étiquettes** :
    - `x_train, y_train = separate_features_labels(train_data)` : Séparation des caractéristiques et étiquettes.
  - **Reshape des caractéristiques** :
    - `x_train = x_train.reshape(-1, img_size, img_size, 1)` : Ajustement de la forme des caractéristiques pour le modèle.

```python
def separate_features_labels(data):
    print("Separating features and labels...")
    features = []
    labels = []
    for img, label in data:
        features.append(img)
        labels.append(label)
    return np.array(features), np.array(labels)

def preprocess_image(image):
    if len(image.shape) == 2:
        image = np.expand_dims(image, axis=-1)
    image = tf.image.resize(image, [img_size, img_size])
    image = image / 255.0  # Normalisation après redimensionnement
    return image

def preprocess_data(data):
    print("Preprocessing data...")
    processed_data = []
    for idx, (img, label) in enumerate(data):
        print(f"Processing image {idx+1}/{len(data)}")
        img = preprocess_image(img)
        processed_data.append([img, label])
    return np.array(processed_data, dtype=object)

# Prétraitement des données
train_data = load_data(data_dir1, data_dir2, limit=100)
train_data = preprocess_data(train_data)
x_train, y_train = separate_features_labels(train_data)
x_train = x_train.reshape(-1, img_size, img_size, 1)
```

### Prétraitement des Données pour l'Entraînement

Ce code est utilisé pour prétraiter les données d'images avant l'entraînement du modèle, en séparant les caractéristiques des étiquettes, redimensionnant et normalisant les images, puis en adaptant les données pour un modèle qui attend des images en trois canaux.

- **Fonctions** :

  - **`separate_features_labels(data)`** :
    - Sépare les caractéristiques (images) des étiquettes dans les données fournies.
    - **Paramètre** : `data` - Données sous forme de liste d'images et d'étiquettes.
    - **Retour** : Deux tableaux NumPy, un pour les caractéristiques et un pour les étiquettes.

  - **`preprocess_image(image)`** :
    - Prétraite une image en la redimensionnant à une taille fixe et en normalisant les valeurs des pixels.
    - **Paramètre** : `image` - Image à prétraiter.
    - **Retour** : Image redimensionnée et normalisée.

  - **`preprocess_data(data)`** :
    - Applique le prétraitement à chaque image des données.
    - **Paramètre** : `data` - Données contenant des images et des étiquettes.
    - **Retour** : Données prétraitées avec des images normalisées et redimensionnées.

- **Application des Fonctions** :
  - **Chargement des données** :
    - `train_data = load_data(data_dir1, data_dir2, limit=100)` : Charge les données d'entraînement à partir des répertoires spécifiés.
  - **Prétraitement des données** :
    - `train_data = preprocess_data(train_data)` : Prétraite les images et les étiquettes.
  - **Séparation des caractéristiques et des étiquettes** :
    - `x_train, y_train = separate_features_labels(train_data)` : Sépare les caractéristiques des étiquettes.
  - **Ajustement des caractéristiques** :
    - `x_train = x_train.reshape(-1, img_size, img_size, 1)` : Redimensionne les images pour ajouter un canal de profondeur.
    - `x_train = np.concatenate([x_train, x_train, x_train], axis=-1)` : Convertit les images en format RGB (3 canaux).


```python
# Séparation des caractéristiques et des étiquettes
def separate_features_labels(data):
    print("Separating features and labels...")
    features = []
    labels = []
    for img, label in data:
        features.append(img)
        labels.append(label)
    return np.array(features), np.array(labels)

# Prétraitement des images
def preprocess_image(image):
    if len(image.shape) == 2:
        image = np.expand_dims(image, axis=-1)
    image = tf.image.resize(image, [img_size, img_size])
    image = image / 255.0  # Normalisation après redimensionnement
    return image

def preprocess_data(data):
    print("Preprocessing data...")
    processed_data = []
    for idx, (img, label) in enumerate(data):
        print(f"Processing image {idx+1}/{len(data)}")
        img = preprocess_image(img)
        processed_data.append([img, label])
    return np.array(processed_data, dtype=object)

# Prétraitement des données
train_data = load_data(data_dir1, data_dir2, limit=100)
train_data = preprocess_data(train_data)
x_train, y_train = separate_features_labels(train_data)
x_train = x_train.reshape(-1, img_size, img_size, 1)
x_train = np.concatenate([x_train, x_train, x_train], axis=-1)  # Convertir en 3 canaux
```

### Préparation des Données pour l'Entraînement

Le code ci-dessous divise les données d'entraînement en ensembles d'entraînement et de validation, puis initialise et ajuste un générateur d'images pour la augmentation des données.

- **Diviser les Données** :
  - **`train_test_split(x_train, x_val, y_train, y_val)`** :
    - **Paramètres** :
      - `x_train` : Caractéristiques des images d'entraînement.
      - `y_train` : Étiquettes des images d'entraînement.
      - `test_size=0.2` : Fraction des données utilisées pour l'ensemble de validation (20%).
      - `random_state=42` : Pour assurer la reproductibilité des splits.
    - **Retour** :
      - `x_train` : Caractéristiques des images d'entraînement (80% des données).
      - `x_val` : Caractéristiques des images de validation (20% des données).
      - `y_train` : Étiquettes des images d'entraînement.
      - `y_val` : Étiquettes des images de validation.

- **Initialiser `ImageDataGenerator`** :
  - **`datagen = ImageDataGenerator(...)`** :
    - **Paramètres** :
      - `width_shift_range=0.2` : Décalage horizontal aléatoire jusqu'à 20% de la largeur de l'image.
      - `height_shift_range=0.2` : Décalage vertical aléatoire jusqu'à 20% de la hauteur de l'image.
      - `horizontal_flip=True` : Retourner horizontalement les images avec une probabilité de 50%.
      - `rotation_range=20` : Rotation aléatoire des images jusqu'à 20 degrés.
      - `zoom_range=0.1` : Zoom aléatoire jusqu'à 10%.
      - `shear_range=0.1` : Déformation par cisaillement jusqu'à 10%.

- **Ajuster `ImageDataGenerator` sur `x_train`** :
  - **`datagen.fit(x_train)`** :
    - Ajuste le générateur d'images avec les caractéristiques d'entraînement pour appliquer les transformations définies.


```python
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)

# Initialiser ImageDataGenerator
print("Initializing ImageDataGenerator...")
datagen = ImageDataGenerator(
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    rotation_range=20,
    zoom_range=0.1,
    shear_range=0.1
)

# Fitter ImageDataGenerator sur x_train
print("Fitting ImageDataGenerator on x_train...")
datagen.fit(x_train)
print("ImageDataGenerator fitting complete.")
```

### Affichage des Images Augmentées

Le code suivant génère et affiche des images augmentées à l'aide du générateur d'images (`ImageDataGenerator`) pour visualiser les transformations appliquées aux images d'entraînement.

- **Fonction `display_augmented_images(datagen, x_train, y_train)`** :
  - **Paramètres** :
    - `datagen` : Instance de `ImageDataGenerator` configurée pour l'augmentation des données.
    - `x_train` : Caractéristiques des images d'entraînement.
    - `y_train` : Étiquettes des images d'entraînement.
  - **Description** :
    - Génère un lot d'images augmentées et les affiche à l'aide de `matplotlib`.
    - Utilise `datagen.flow(x_train, y_train, batch_size=5)` pour obtenir un lot de 5 images augmentées.
    - Affiche les images dans une grille sans axes.

- **Exécution** :
  - **`display_augmented_images(datagen, x_train, y_train)`** :
    - Appelle la fonction pour afficher les images augmentées générées à partir de `x_train` et `y_train`.

```python
def display_augmented_images(datagen, x_train, y_train):
    print("Generating augmented images...")
    plt.figure(figsize=(10, 10))
    for x_batch, y_batch in datagen.flow(x_train, y_train, batch_size=5):
        for i in range(5):
            plt.subplot(330 + 1 + i)
            plt.xticks([])
            plt.yticks([])
            plt.imshow(x_batch[i].reshape(img_size, img_size, 3), cmap='gray')
        plt.show()
        break

display_augmented_images(datagen, x_train, y_train)
```

### Définition et Compilation du Modèle

Ce code définit un modèle de classification d'images en utilisant le réseau de neurones pré-entraîné VGG16 comme base, ajoute des couches personnalisées, et compile le modèle pour l'entraînement.

- **Définition du Modèle** :
  - **Base Model** :
    - **`base_model = tf.keras.applications.VGG16(...)`** :
      - Utilise le modèle pré-entraîné VGG16, sans les couches de classification supérieures (`include_top=False`), avec les poids pré-entraînés sur ImageNet.
      - **Paramètres** :
        - `input_shape=(img_size, img_size, 3)` : Spécifie la forme des images d'entrée (taille des images et nombre de canaux).
        - `weights='imagenet'` : Charge les poids pré-entraînés de ImageNet.
      - **`base_model.trainable = False`** : Gèle les couches du modèle de base pour éviter leur mise à jour pendant l'entraînement.
  
  - **Modèle Personnalisé** :
    - **`model = Sequential([...])`** :
      - **`base_model`** : Ajoute le modèle pré-entraîné comme première couche.
      - **`Flatten()`** : Aplatisse les sorties du modèle de base pour les passer aux couches denses.
      - **`Dense(128, activation='relu')`** : Ajoute une couche dense avec 128 unités et activation ReLU.
      - **`Dropout(0.5)`** : Applique un dropout de 50% pour éviter le surapprentissage.
      - **`Dense(1, activation='sigmoid')`** : Ajoute une couche de sortie avec une seule unité et activation sigmoïde pour la classification binaire.

- **Compilation du Modèle** :
  - **`model.compile(...)`** :
    - **Paramètres** :
      - `optimizer=tf.keras.optimizers.Adam(learning_rate=0.001)` : Utilise l'optimiseur Adam avec un taux d'apprentissage de 0.001.
      - `loss='binary_crossentropy'` : Fonction de perte pour la classification binaire.
      - `metrics=['accuracy']` : Métriques à suivre pendant l'entraînement.
  
  - **Forcer la Construction** :
    - **`model.build(input_shape=(None, img_size, img_size, 3))`** : Force la construction du modèle avec une forme d'entrée spécifiée.

- **Affichage du Résumé du Modèle** :
  - **`model.summary()`** : Affiche un résumé du modèle, y compris les couches, les formes de sortie, et le nombre de paramètres.


```python
print("Defining the model...")

base_model = tf.keras.applications.VGG16(input_shape=(img_size, img_size, 3),
                                         include_top=False,
                                         weights='imagenet')
base_model.trainable = False  # Geler les couches du modèle de base

model = Sequential([
    base_model,
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

print("Model defined.")
print("Compiling the model...")

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), 
              loss='binary_crossentropy', 
              metrics=['accuracy'])

# Forcer la construction du modèle
model.build(input_shape=(None, img_size, img_size, 3))

# Affichage du résumé du modèle
print("Model compiled.")
print("Model summary:")
print(model.summary())
```

### Configuration des Callbacks et Entraînement du Modèle

Ce code configure les callbacks pour ajuster le taux d'apprentissage et arrêter l'entraînement de manière précoce si nécessaire, puis entraîne le modèle avec ces configurations.

- **Callbacks** :
  - **`ReduceLROnPlateau`** :
    - **`monitor='val_loss'`** : Surveille la perte de validation pour décider quand ajuster le taux d'apprentissage.
    - **`factor=0.5`** : Réduit le taux d'apprentissage de moitié lorsqu'une stagnation est détectée.
    - **`patience=3`** : Nombre d'époques à attendre avant de réduire le taux d'apprentissage si la perte de validation ne s'améliore pas.
    - **`min_lr=0.0001`** : Taux d'apprentissage minimum auquel il peut être réduit.
    
  - **`EarlyStopping`** :
    - **`monitor='val_loss'`** : Surveille la perte de validation pour décider quand arrêter l'entraînement.
    - **`patience=5`** : Nombre d'époques à attendre avant d'arrêter si la perte de validation ne s'améliore pas.
    - **`restore_best_weights=True`** : Restaure les poids du modèle aux meilleurs poids observés pendant l'entraînement.

- **Entraînement du Modèle** :
  - **`model.fit(...)`** :
    - **`datagen.flow(x_train, y_train, batch_size=batch_size)`** : Génère des lots d'images augmentées pour l'entraînement.
    - **`epochs=epochs`** : Nombre total d'époques pour l'entraînement.
    - **`validation_data=(x_val, y_val)`** : Ensemble de validation pour évaluer les performances du modèle pendant l'entraînement.
    - **`callbacks=[reduce_lr, early_stop]`** : Liste des callbacks à utiliser pendant l'entraînement.

- **Exécution** :
  - **`print("Starting training...")`** : Annonce le début de l'entraînement.
  - **`print("Training complete.")`** : Annonce la fin de l'entraînement.


```python
# Callbacks
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=0.0001)
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Entraînement du modèle
print("Starting training...")
history = model.fit(datagen.flow(x_train, y_train, batch_size=batch_size),
                    epochs=epochs,
                    validation_data=(x_val, y_val),
                    callbacks=[reduce_lr, early_stop])
print("Training complete.")
```

### Tracé des Courbes d'Apprentissage

Le code ci-dessous génère et affiche des courbes d'apprentissage pour la perte et la précision, en utilisant l'historique de l'entraînement du modèle.

- **Fonction `plot_learning_curves(history)`** :
  - **Paramètre** :
    - `history` : Historique de l'entraînement du modèle, retourné par `model.fit`.
  - **Description** :
    - **Courbe de Perte** :
      - Affiche la perte d'entraînement et la perte de validation en fonction du nombre d'époques.
      - **Axes** :
        - **X** : Nombre d'époques.
        - **Y** : Perte.
    - **Courbe de Précision** :
      - Affiche la précision d'entraînement et la précision de validation en fonction du nombre d'époques.
      - **Axes** :
        - **X** : Nombre d'époques.
        - **Y** : Précision.
    - **Disposition** :
      - Utilise `plt.tight_layout()` pour optimiser l'agencement des sous-graphiques.

- **Exécution** :
  - **`plot_learning_curves(history)`** : Appelle la fonction pour afficher les courbes d'apprentissage.

```python
def plot_learning_curves(history):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()

plot_learning_curves(history)
```

### Évaluation du Modèle

Le code ci-dessous évalue le modèle sur l'ensemble de validation et affiche les métriques de performance : la précision et la perte de validation.

- **Évaluation du Modèle** :
  - **`model.evaluate(x_val, y_val, verbose=0)`** :
    - **`x_val`** : Les caractéristiques de l'ensemble de validation.

  - **Résultats** :

### Courbe ROC et AUC

Le code ci-dessous calcule et affiche la courbe ROC (Receiver Operating Characteristic) ainsi que l'AUC (Area Under the Curve) pour les prédictions du modèle sur l'ensemble d'entraînement.

- **Calcul de la Courbe ROC** :
  - **`roc_curve(y_train, y_pred)`** :
    - **`y_train`** : Étiquettes réelles des images d'entraînement.
    - **`y_pred`** : Étiquettes prédites par le modèle, transformées en probabilités.
    - Renvoie les taux de faux positifs (FPR) et les taux de vrais positifs (TPR) pour différentes valeurs de seuil.
  - **`auc(fpr, tpr)`** :
    - Calcule l'aire sous la courbe ROC (AUC) en utilisant les valeurs FPR et TPR.

- **Affichage de la Courbe ROC** :
  - **`plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)`** :
    - Trace la courbe ROC avec une couleur orange foncé.
    - Affiche l'aire sous la courbe (AUC) dans la légende.
  - **`plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')`** :
    - Trace une ligne diagonale représentant le modèle aléatoire (référence).
  - **Axes et Titre** :
    - **`plt.xlim([0.0, 1.0])`** et **`plt.ylim([0.0, 1.05])`** : Limites des axes.
    - **`plt.xlabel('False Positive Rate')`** et **`plt.ylabel('True Positive Rate')`** : Étiquettes des axes.
    - **`plt.title('Receiver Operating Characteristic')`** : Titre du graphique.
  - **`plt.legend(loc="lower right")`** : Positionne la légende en bas à droite.

```python
fpr, tpr, _ = roc_curve(y_train, y_pred)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

### Affichage des Prédictions Simples

Le code ci-dessous génère et affiche un certain nombre d'images avec leurs prédictions et labels réels pour évaluer visuellement les performances du modèle.

- **Fonction `display_predictions_simple`** :
  - **Arguments** :
    - **`model`** : Le modèle de classification pour effectuer les prédictions.
    - **`x_train`** : Les images d'entraînement utilisées pour les prédictions.
    - **`y_train`** : Les étiquettes réelles des images d'entraînement.
    - **`num_images`** : Nombre d'images à afficher (défini ici à 5).

  - **Prédictions et Affichage** :
    - **`predictions = (model.predict(x_train) > 0.5).astype("int32")`** :
      - Effectue des prédictions sur les images d'entraînement et convertit les probabilités en étiquettes binaires (0 ou 1).
    - **`plt.figure(figsize=(10, 10))`** :
      - Crée une figure pour afficher les images.
    - **`for i in range(num_images)`** :
      - Boucle pour afficher le nombre d'images spécifié.
      - **`plt.subplot(1, num_images, i + 1)`** : Crée une sous-figure pour chaque image.
      - **`plt.xticks([])`** et **`plt.yticks([])`** : Supprime les ticks des axes pour une présentation plus claire.
      - **`plt.imshow(x_train[i].reshape(img_size, img_size, 3))`** : Affiche l'image redimensionnée.
      - **`color = 'green' if pred_label == true_label else 'red'`** : Détermine la couleur du titre en fonction de la précision de la prédiction.
      - **`plt.title(f"Pred: {pred_label}, True: {true_label}", color=color)`** : Affiche le titre avec la couleur appropriée (vert pour les bonnes prédictions, rouge pour les erreurs).
    - **`plt.show()`** : Affiche le graphique.

```python
def display_predictions_simple(model, x_train, y_train, num_images=5):
    predictions = (model.predict(x_train) > 0.5).astype("int32")
    plt.figure(figsize=(10, 10))
    for i in range(num_images):
        plt.subplot(1, num_images, i + 1)
        plt.xticks([])
        plt.yticks([])
        plt.imshow(x_train[i].reshape(img_size, img_size, 3))
        pred_label = predictions[i][0]
        true_label = y_train[i]
        color = 'green' if pred_label == true_label else 'red'
        plt.title(f"Pred: {pred_label}, True: {true_label}", color=color)
    plt.show()

display_predictions_simple(model, x_train, y_train, num_images=5)
```

### Affichage de l'Histogramme des Prédictions

Le code ci-dessous permet d'obtenir des statistiques descriptives sur les prédictions faites par le modèle et affiche un histogramme de la distribution des probabilités prédites.

- **Fonction `plot_prediction_histogram`** :
  - **Arguments** :
    - **`model`** : Le modèle de classification utilisé pour générer les prédictions.
    - **`x_data`** : Les données sur lesquelles les prédictions sont effectuées.

  - **Procédure** :
    1. **Obtention des Prédictions** :
       - **`predictions = model.predict(x_data).flatten()`** :
         - Effectue les prédictions sur les données d'entrée et les aplatit en un vecteur 1D.
    2. **Affichage des Statistiques de Base** :
       - **`print(f"Nombre total de prédictions: {len(predictions)}")`** :
         - Affiche le nombre total de prédictions.
       - **`print(f"Valeur minimale des prédictions: {predictions.min():.4f}")`** :
         - Affiche la valeur minimale des prédictions.
       - **`print(f"Valeur maximale des prédictions: {predictions.max():.4f}")`** :
         - Affiche la valeur maximale des prédictions.
       - **`print(f"Moyenne des prédictions: {predictions.mean():.4f}")`** :
         - Affiche la moyenne des prédictions.
       - **`print(f"Médiane des prédictions: {np.median(predictions):.4f}")`** :
         - Affiche la médiane des prédictions.
       - **`print(f"Écart-type des prédictions: {predictions.std():.4f}")`** :
         - Affiche l'écart-type des prédictions.
    3. **Calcul et Affichage des Bins de l'Histogramme** :
       - **`counts, bin_edges = np.histogram(predictions, bins=20)`** :
         - Calcule les fréquences des prédictions pour chaque intervalle (bin).
       - **`for i in range(len(counts))`** :
         - Affiche les bornes et le nombre d'images pour chaque bin.
    4. **Affichage de l'Histogramme** :
       - **`plt.hist(predictions, bins=20, color='blue', alpha=0.7)`** :
         - Crée un histogramme de la distribution des prédictions.
       - **`plt.xlabel('Probabilité prédite')`** :
         - Étiquette de l'axe des x.
       - **`plt.ylabel('Nombre d\'images')`** :
         - Étiquette de l'axe des y.
       - **`plt.title('Distribution des Prédictions')`** :
         - Titre du graphique.
       - **`plt.show()`** :
         - Affiche le graphique.

```python
def plot_prediction_histogram(model, x_data):
    # Obtenir les prédictions
    predictions = model.predict(x_data).flatten()

    # Affichage des statistiques de base
    print(f"Nombre total de prédictions: {len(predictions)}")
    print(f"Valeur minimale des prédictions: {predictions.min():.4f}")
    print(f"Valeur maximale des prédictions: {predictions.max():.4f}")
    print(f"Moyenne des prédictions: {predictions.mean():.4f}")
    print(f"Médiane des prédictions: {np.median(predictions):.4f}")
    print(f"Écart-type des prédictions: {predictions.std():.4f}")

    # Calcul et affichage des bins de l'histogramme
    counts, bin_edges = np.histogram(predictions, bins=20)
    for i in range(len(counts)):
        print(f"Bin {i+1}: {bin_edges[i]:.2f} - {bin_edges[i+1]:.2f}, Nombre d'images: {counts[i]}")

    # Affichage de l'histogramme
    plt.hist(predictions, bins=20, color='blue', alpha=0.7)
    plt.xlabel('Probabilité prédite')
    plt.ylabel('Nombre d\'images')
    plt.title('Distribution des Prédictions')
    plt.show()
    
plot_prediction_histogram(model, x_train)
```

### Tableau de prédiction

Cette fonction affiche un tableau des prédictions faites par le modèle pour un échantillon de données. Elle inclut les informations suivantes :

- **Index** : L'indice de l'échantillon.
- **Prédiction** : La classe prédite par le modèle.
- **Probabilité** : La probabilité associée à la prédiction.
- **Vrai Label** : L'étiquette vraie de l'échantillon.
- **Correct** : Indique si la prédiction est correcte (Oui/Non).


```python
def print_predictions_table(model, x_data, y_data, num_samples=10):
    probs = model.predict(x_data).flatten()
    predictions = (probs > 0.5).astype("int32")
    print(f"{'Index':<6}{'Prédiction':<12}{'Probabilité':<12}{'Vrai Label':<12}{'Correct':<8}")
    print("-" * 60)
    for i in range(num_samples):
        correct = 'Oui' if predictions[i] == y_data[i] else 'Non'
        print(f"{i:<6}{predictions[i]:<12}{probs[i]:<12.4f}{y_data[i]:<12}{correct:<8}")

print_predictions_table(model, x_train, y_train)
```
