## Objectif

Ce notebook a pour objectif d'explorer l'effet de la technique de **Data Augmentation par recadrage (Crop)** dans l'entraînement d'un modèle de réseau de neurones convolutifs (CNN) pour la classification d'images. Cette approche permet d'augmenter la diversité des données d'entraînement, améliorant ainsi la robustesse et la généralisation du modèle. L'ensemble de données utilisé pour cette étude est le **CIFAR-10**, contenant 10 classes d'objets variés.

## Étapes de Démarche

1. **Vérification de la Disponibilité du GPU**
   - Validation de la présence d'un GPU pour accélérer l'entraînement du modèle.
2. **Chargement et Préparation des Données**
   - Chargement du dataset CIFAR-10, redimensionnement des images, conversion en niveaux de gris, normalisation des valeurs des pixels, et restructuration des images pour les adapter à l'entrée du modèle.
3. **Visualisation des Données**
   - Aperçu visuel des images avant augmentation, permettant de comprendre les données de base.
4. **Augmentation des Données**
   - Application de la technique de recadrage central, en combinaison avec d'autres transformations comme le zoom, la translation et la rotation, pour augmenter la variabilité des données d'entraînement.
5. **Définition et Entraînement du Modèle CNN**
   - Construction d'un modèle CNN comprenant des couches de convolution, de pooling, et des couches fully connected. Entraînement du modèle avec les données augmentées.
6. **Évaluation du Modèle**
   - Test du modèle sur les données de validation pour évaluer sa précision et sa capacité de généralisation.
7. **Visualisation des Performances**
   - Analyse approfondie des performances du modèle à l'aide de différents outils :
     - **Matrice de confusion** : Visualisation des erreurs de classification.
     - **Rapport de classification** : Evaluation des performances pour chaque classe.
     - **Courbe ROC** : Analyse de la capacité du modèle à distinguer les différentes classes à divers seuils.
8. **Heatmaps des Activations**
   - Visualisation des activations des couches internes du modèle pour interpréter les caractéristiques apprises.

## Interprétation et Résultats

- **Courbes d'Apprentissage** : Analyse de la convergence du modèle et de la relation entre la perte (loss) et l'exactitude (accuracy) au cours des époques d'entraînement.
- **Matrice de Confusion** : Examen détaillé des prédictions correctes et incorrectes du modèle, permettant de mieux comprendre où le modèle se trompe.
- **Rapport de Classification** : Evaluation quantitative des performances du modèle en termes de précision, de rappel et de score F1 pour chaque classe.
- **Courbe ROC** : Evaluation de la capacité du modèle à distinguer les classes dans des scénarios de classification binaire et multiclasses.
- **Heatmaps des Activations** : Exploration des cartes de caractéristiques extraites par les couches convolutives du modèle, offrant une meilleure compréhension des régions d'intérêt dans les images.

```python
TRAIN_FOLDER = '../../../chest_xray/train'
TEST_FOLDER = '../../../chest_xray/test'
VAL_FOLDER = '../../../chest_xray/val'
print("folders")
```

```python
img_size = 150
img_width, img_height = 150, 150
classes = ['PNEUMONIA', 'NORMAL']

def load_data(folders):
    data = []  # Initialisation d'une liste qui va stocker les images et les étiquettes (PNEUMONIA ou NORMAL)
    labels = ['PNEUMONIA', 'NORMAL']  # Liste des labels possibles
    
    # Pour chaque répertoire dans la liste des répertoires
    for folder in folders:
        for label in labels:  # Pour chaque label dans les répertoires
            path = os.path.join(folder, label)  # Construction du chemin
            # On vérifie si le chemin existe
            if not os.path.exists(path):
                print(f"Le répertoire {path} n'existe pas. Passer au suivant.")
                continue
            class_num = labels.index(label)  # Conversion du nom de la classe en numéro
            
            # Parcourir tous les fichiers du répertoire
            for img in os.listdir(path):
                try:
                    img_arr = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)  # Lire l'image en niveau de gris
                    if img_arr is None:
                        print(f"Failed to read {img}. Skipping.")
                        continue
                    resized_arr = cv2.resize(img_arr, (img_size, img_size))  # Redimensionner l'image
                    data.append([resized_arr, class_num])  # Ajouter l'image redimensionnée et son label à la liste
                except Exception as e:
                    print(f"Erreur avec l'image {img}: {e}")
    return np.array(data, dtype=object)  # Convertir la liste en tableau NumPy et retourner

# Charger les données à partir des trois répertoires
folders = [TRAIN_FOLDER, TEST_FOLDER, VAL_FOLDER]
data = load_data(folders)

print("Données chargées avec succès")
```

```python
def make_square(img):
    size = min(img.shape[0], img.shape[1])
    return cv2.resize(img[:size, :size], (img_width, img_height))

print("make square succès")
```

```python
def generate_cropped_images(folder, img_width, img_height, classes):
    cropped_images = []
    labels = []
    for label in classes:
        path = os.path.join(folder, label)
        if not os.path.exists(path):
            print(f"Le répertoire {path} n'existe pas. Passer au suivant.")
            continue
        for filename in os.listdir(path):
            img_path = os.path.join(path, filename)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                cropped_img = make_square(img)
                cropped_images.append(cropped_img)
                labels.append(classes.index(label))
            else:
                print(f"Impossible de lire l'image {img_path}")
    return np.array(cropped_images), np.array(labels)

print("generating cropped images")
```

```python
def generate_augmented_images(images, labels):
    datagen = ImageDataGenerator(
        rotation_range=15,
        width_shift_range=0.1,
        height_shift_range=0.1,
        shear_range=0.1,
        zoom_range=0.1,
        horizontal_flip=True,
        fill_mode='nearest'
    )
    datagen.fit(images)
    return datagen

print("generating augmented images")
```

```python
# Génération des images recadrées
x_train, y_train = generate_cropped_images(TRAIN_FOLDER, img_width, img_height, classes)
print("Images d'entraînement générées avec succès.")

x_test, y_test = generate_cropped_images(TEST_FOLDER, img_width, img_height, classes)
print("Images de test générées avec succès.")

x_val, y_val = generate_cropped_images(VAL_FOLDER, img_width, img_height, classes)
print("Images de validation générées avec succès.")

# Normalisation des images
x_train = x_train.reshape(-1, img_width, img_height, 1) / 255.0
print("Images d'entraînement normalisées avec succès.")

x_test = x_test.reshape(-1, img_width, img_height, 1) / 255.0
print("Images de test normalisées avec succès.")

x_val = x_val.reshape(-1, img_width, img_height, 1) / 255.0
print("Images de validation normalisées avec succès.")

# Conversion des labels en format one-hot encoding
y_train = tf.keras.utils.to_categorical(y_train, num_classes=2)
print("Labels d'entraînement convertis en one-hot encoding avec succès.")

y_test = tf.keras.utils.to_categorical(y_test, num_classes=2)
print("Labels de test convertis en one-hot encoding avec succès.")

y_val = tf.keras.utils.to_categorical(y_val, num_classes=2)
print("Labels de validation convertis en one-hot encoding avec succès.")
```

```python
def plot_class_distribution(y_train, classes):
    # Calcul des classes uniques et de leurs comptes
    unique, counts = np.unique(np.argmax(y_train, axis=1), return_counts=True)

    print("Classes uniques:", unique)
    print("Comptes des classes:", counts)

    plt.bar(classes, counts) # Tracé de la distribution des classes
    plt.title('Distribution des classes')
    plt.show()

# Affichage de la distribution des classes
plot_class_distribution(y_train, classes)
print("Fonction plot_class_distribution exécutée avec succès.")
```

```python
def create_cnn_model(img_width, img_height):
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(img_width, img_height, 1)),
        MaxPooling2D((2, 2)),
        Dropout(0.2),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Dropout(0.2),
        Conv2D(128, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Dropout(0.2),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.5),
        Dense(2, activation='softmax')
    ])
    
    # Compile the model
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    model.summary()
    print("Résumé du modèle affiché avec succès.")

    
    return model

# Exemple d'utilisation de la fonction
img_width, img_height = 150, 150  # Assurez-vous que ces dimensions correspondent à vos besoins architecturaux

# Création du modèle
model = create_cnn_model(img_width, img_height)
print("Modèle créé et compilé avec succès.")
```

```python
def create_model(img_width, img_height):
    model = Sequential()
    print("Modèle créé.")

    model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(img_width, img_height, 1)))
    print("Couche Conv2D ajoutée avec 32 filtres.")
    model.add(MaxPooling2D(pool_size=(2, 2)))
    print("Couche MaxPooling2D ajoutée.")

    model.add(Conv2D(64, (3, 3), activation='relu'))
    print("Couche Conv2D ajoutée avec 64 filtres.")
    model.add(MaxPooling2D(pool_size=(2, 2)))
    print("Couche MaxPooling2D ajoutée.")

    model.add(Conv2D(128, (3, 3), activation='relu'))
    print("Couche Conv2D ajoutée avec 128 filtres.")
    model.add(MaxPooling2D(pool_size=(2, 2)))
    print("Couche MaxPooling2D ajoutée.")

    model.add(Conv2D(256, (3, 3), activation='relu'))
    print("Couche Conv2D ajoutée avec 256 filtres.")
    model.add(MaxPooling2D(pool_size=(2, 2)))
    print("Couche MaxPooling2D ajoutée.")

    model.add(Flatten())
    print("Couche Flatten ajoutée.")

    model.add(Dense(256, activation='relu'))
    print("Couche Dense ajoutée avec 256 unités.")
    model.add(Dropout(0.5))
    print("Couche Dropout ajoutée avec un taux de 0.5.")
    model.add(Dense(128, activation='relu'))
    print("Couche Dense ajoutée avec 128 unités.")
    model.add(Dropout(0.5))
    print("Couche Dropout ajoutée avec un taux de 0.5.")
    model.add(Dense(2, activation='softmax'))
    print("Couche Dense ajoutée avec 2 unités et activation softmax.")

    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    print("Modèle compilé avec optimiseur 'adam', perte 'categorical_crossentropy' et métrique 'accuracy'.")

    model.summary()
    print("Résumé du modèle affiché.")

    return model
```

```python
def cross_validate_model(x_train, y_train, img_width, img_height, k_folds=5):
    kf = KFold(n_splits=k_folds, shuffle=True)
    fold_accuracies = []
    fold_losses = []
    val_accuracies = []
    val_losses = []
    
    for fold, (train_idx, val_idx) in enumerate(kf.split(x_train)):
        print(f"Création du modèle pour le fold {fold + 1}...")
        model = create_model(img_width, img_height)
        print("Modèle créé et compilé avec succès.")

        # Entraînement du modèle
        history = model.fit(x_train[train_idx], y_train[train_idx], epochs=10, batch_size=32, 
                            validation_data=(x_train[val_idx], y_train[val_idx]), verbose=1)

        # Stockage des performances
        val_accuracy = history.history['val_accuracy']
        val_loss = history.history['val_loss']
        val_accuracies.extend(val_accuracy)
        val_losses.extend(val_loss)
        
        # Évaluation du modèle
        val_loss, val_accuracy = model.evaluate(x_train[val_idx], y_train[val_idx], verbose=0)
        fold_accuracies.append(val_accuracy * 100)
        fold_losses.append(val_loss)
        print(f'Fold {fold + 1} accuracy: {val_accuracy * 100:.2f}%')
        print(f'Fold {fold + 1} loss: {val_loss:.4f}')
        
    mean_accuracy = np.mean(fold_accuracies)
    mean_loss = np.mean(fold_losses)
    print(f'\nMean accuracy over {k_folds} folds: {mean_accuracy:.2f}%')
    print(f'Mean loss over {k_folds} folds: {mean_loss:.4f}')
    
    # Plotting accuracies and losses
    plt.figure(figsize=(12, 6))

    # Accuracy plot
    plt.subplot(1, 2, 1)
    plt.plot(val_accuracies, label='Validation Accuracy')
    plt.title('Validation Accuracy per Fold')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    # Loss plot
    plt.subplot(1, 2, 2)
    plt.plot(val_losses, label='Validation Loss')
    plt.title('Validation Loss per Fold')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
cross_validate_model(x_train, y_train, img_width, img_height)
final_model = create_model(img_width, img_height)
final_model.fit(x_train, y_train, epochs=10, batch_size=32)
```

```python
# Fonction pour tracer les performances du modèle final
def plot_final_model_performance(history):
    # Plot loss and accuracy
    plt.figure(figsize=(12, 6)) 

    # Accuracy plot
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    # Loss plot
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Création et entraînement du modèle final
final_model = create_model(img_width, img_height)
history = final_model.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_val, y_val))

# Affichage des performances du modèle final
plot_final_model_performance(history)
```

```python
def evaluate_model_on_test_set(model, x_test, y_test, classes):
    # Prédiction des classes
    y_pred = model.predict(x_test)
    print("Prédictions effectuées avec succès.")

    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true_classes = np.argmax(y_test, axis=1)
    print("Classes prédites et vraies calculées avec succès.")

    # Matrice de confusion
    cm = confusion_matrix(y_true_classes, y_pred_classes)
    print("Matrice de confusion calculée avec succès.")

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.title('Confusion Matrix')
    plt.show()
    print("Matrice de confusion affichée avec succès.")

    # Rapport de classification
    print("Classification Report:\n", classification_report(y_true_classes, y_pred_classes, target_names=classes))
    print("Rapport de classification affiché avec succès.")

    # Courbe ROC
    fpr, tpr, _ = roc_curve(y_true_classes, y_pred[:, 1])
    roc_auc = auc(fpr, tpr)
    print("Courbe ROC et AUC calculées avec succès.")

    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    plt.show()
    print("Courbe ROC affichée avec succès.")

evaluate_model_on_test_set(final_model, x_test, y_test, classes)
```

```python
def make_gradcam_heatmap(img_array, model, layer_name):
    """
    Crée une heatmap Grad-CAM pour une image donnée et un modèle donné.
    """
    grad_model = tf.keras.models.Model(
        inputs=[model.inputs],
        outputs=[model.get_layer(layer_name).output, model.output]
    )

    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        loss = predictions[:, np.argmax(predictions[0])]

    grads = tape.gradient(loss, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    
    conv_outputs = conv_outputs[0]
    conv_outputs = np.maximum(conv_outputs, 0)

    for i in range(pooled_grads.shape[-1]):
        conv_outputs[:, :, i] *= pooled_grads[i]

    heatmap = np.mean(conv_outputs, axis=-1)
    heatmap = np.maximum(heatmap, 0)
    heatmap /= np.max(heatmap)
    
    return heatmap

def display_heatmaps(image_path, heatmap):
    """
    Affiche une image avec une heatmap superposée.
    """
    img = image.load_img(image_path)
    img = image.img_to_array(img)
    
    heatmap_resized = cv2.resize(heatmap, (img.shape[1], img.shape[0]))
    heatmap_resized = np.uint8(255 * heatmap_resized)
    heatmap_resized = cv2.applyColorMap(heatmap_resized, cv2.COLORMAP_JET)
    superimposed_img = cv2.addWeighted(img, 0.6, heatmap_resized, 0.4, 0)
    
    plt.imshow(superimposed_img.astype('uint8'))
    plt.title("Superimposed Heatmap")
    plt.axis('off')
    plt.show()

def generate_heatmaps_for_images(image_paths, model, layer_name, sample_size=5):
    """
    Génère et affiche des heatmaps pour un échantillon d'images.
    """
    img_width, img_height = 224, 224 
    sample_paths = image_paths[:sample_size] 
    for img_path in sample_paths:
        img = image.load_img(img_path, target_size=(img_width, img_height))
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        img_array = img_array / 255.0 
        
        heatmap = make_gradcam_heatmap(img_array, model, layer_name)
        display_heatmaps(img_path, heatmap)

TRAIN_FOLDER = '../../../chest_xray/train'
TEST_FOLDER = '../../../chest_xray/test'
VAL_FOLDER = '../../../chest_xray/val'

image_paths = [os.path.join(TEST_FOLDER, 'NORMAL', img) for img in os.listdir(os.path.join(TEST_FOLDER, 'NORMAL'))][:10] 
last_conv_layer_name = 'conv2d_3' 
final_model = tf.keras.models.load_model('path/to/your/model')
generate_heatmaps_for_images(image_paths, final_model, last_conv_layer_name, sample_size=5)
```
