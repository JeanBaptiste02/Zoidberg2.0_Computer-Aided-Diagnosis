## Objectif

Ce notebook a pour objectif d'explorer l'effet de la technique de Data Augmentation par recadrage (Crop) et retournement horizontal (Flip) dans l'entraînement d'un modèle de réseau de neurones convolutifs (CNN) pour la classification d'images. Ces techniques permettent d'augmenter la diversité des données d'entraînement, ce qui améliore la robustesse et la généralisation du modèle. L'ensemble de données utilisé pour cette étude est un dataset de rayons X pour la détection de pathologies pulmonaires, structuré en trois ensembles : entraînement, validation, et test.

## Étapes de Démarche

1. **Vérification de la Disponibilité du GPU**
   - Validation de la présence d'un GPU pour accélérer l'entraînement du modèle.
2. **Chargement et Préparation des Données**
   - Chargement du dataset CIFAR-10, redimensionnement des images, conversion en niveaux de gris, normalisation des valeurs des pixels, et restructuration des images pour les adapter à l'entrée du modèle.
3. **Visualisation des Données**
   - Aperçu visuel des images avant augmentation, permettant de comprendre les données de base.
4. **Augmentation des Données**
   - Application de la technique de recadrage central, en combinaison avec d'autres transformations comme le zoom, la translation et la rotation, pour augmenter la variabilité des données d'entraînement.
5. **Définition et Entraînement du Modèle CNN**
   - Construction d'un modèle CNN comprenant des couches de convolution, de pooling, et des couches fully connected. Entraînement du modèle avec les données augmentées.
6. **Évaluation du Modèle**
   - Test du modèle sur les données de validation pour évaluer sa précision et sa capacité de généralisation.
7. **Visualisation des Performances**
   - Analyse approfondie des performances du modèle à l'aide de différents outils :
     - **Matrice de confusion** : Visualisation des erreurs de classification.
     - **Rapport de classification** : Evaluation des performances pour chaque classe.
     - **Courbe ROC** : Analyse de la capacité du modèle à distinguer les différentes classes à divers seuils.
8. **Heatmaps des Activations**
   - Visualisation des activations des couches internes du modèle pour interpréter les caractéristiques apprises.

## Interprétation et Résultats

- **Courbes d'Apprentissage** : Analyse de la convergence du modèle et de la relation entre la perte (loss) et l'exactitude (accuracy) au cours des époques d'entraînement.

### Définition des chemins des dossiers et affichage d'un message

Le code suivant définit les chemins d'accès aux dossiers contenant les données pour l'entraînement, les tests et la validation d'un modèle. Ensuite, il affiche un message pour indiquer que les chemins des dossiers sont configurés.


```python
TRAIN_FOLDER = '../../../chest_xray/train'
TEST_FOLDER = '../../../chest_xray/test'
VAL_FOLDER = '../../../chest_xray/val'
print("folders")
```

### Chargement et Prétraitement des Images

Le code suivant définit une fonction pour charger les images à partir d'un dossier, les redimensionner, et les convertir en tableaux NumPy. Ensuite, il utilise cette fonction pour charger les images d'entraînement, de test, et de validation, et affiche les formes des tableaux résultants.

- **Fonction `load_images_from_folder`** :
  - **Paramètres** :
    - `folder` : Chemin du dossier contenant les sous-dossiers de classes d'images.
    - `size` : Taille de redimensionnement des images (par défaut `(32, 32)`).
  - **Retour** :
    - Deux tableaux NumPy : un pour les images et un pour les étiquettes.

- **Chargement des données** :
  - Les images sont chargées depuis les dossiers spécifiés pour l'entraînement, le test, et la validation.
  - Les formes des tableaux d'images sont affichées pour vérifier que les données ont été chargées correctement.


```python
def load_images_from_folder(folder, size=(32, 32)):
    images = []
    labels = []
    class_names = sorted(os.listdir(folder))  # Assume each subfolder is a class
    for label, class_name in enumerate(class_names):
        class_folder = os.path.join(folder, class_name)
        for filename in os.listdir(class_folder):
            img_path = os.path.join(class_folder, filename)
            img = tf.keras.preprocessing.image.load_img(img_path, target_size=size)
            img = tf.keras.preprocessing.image.img_to_array(img)
            images.append(img)
            labels.append(label)
    return np.array(images), np.array(labels)

# Charger et prétraiter les données
print("Chargement des données...")
train_images, train_labels = load_images_from_folder(TRAIN_FOLDER)
test_images, test_labels = load_images_from_folder(TEST_FOLDER)
val_images, val_labels = load_images_from_folder(VAL_FOLDER)

print(f"Train images shape: {train_images.shape}")
print(f"Test images shape: {test_images.shape}")
print(f"Validation images shape: {val_images.shape}")
```

### Normalisation des Images

Le code suivant normalise les images en divisant les valeurs des pixels par 255.0. Cela transforme les valeurs des pixels de la plage `[0, 255]` à la plage `[0, 1]`. Ensuite, il affiche les formes des tableaux d'images après normalisation pour vérifier que la transformation a été effectuée correctement.

- **Normalisation** :
  - Les images d'entraînement, de test, et de validation sont normalisées en divisant les valeurs des pixels par 255.0.

- **Affichage des formes après normalisation** :
  - Les formes des tableaux d'images après la normalisation sont affichées pour vérifier que les données ont été correctement transformées.

```python
train_images, test_images, val_images = train_images / 255.0, test_images / 255.0, val_images / 255.0
print(f"Train images shape (après normalisation): {train_images.shape}")
print(f"Test images shape (après normalisation): {test_images.shape}")
print(f"Validation images shape (après normalisation): {val_images.shape}")
```

### Augmentation des Données et Visualisation

Le code suivant applique des techniques d'augmentation des données, comme le recadrage aléatoire et le retournement horizontal, sur les images d'entraînement, de test, et de validation. Il affiche également les images originales ainsi que les images augmentées pour vérifier les transformations appliquées.

- **Définition de l'augmentation des données** :
  - `data_augmentation` : Séquence d'augmentation des données utilisant le recadrage aléatoire (`RandomCrop`) et le retournement horizontal (`RandomFlip`).

- **Affichage des images originales** :
  - La fonction `plot_images` affiche un échantillon d'images originales avec leurs étiquettes de classe.

- **Application de l'augmentation des données** :
  - Les images d'entraînement, de test, et de validation sont transformées par la séquence d'augmentation définie.

- **Affichage des formes après augmentation** :
  - Les formes des tableaux d'images après augmentation sont affichées pour vérifier que la transformation a été correctement appliquée.

- **Affichage des images augmentées** :
  - La fonction `plot_augmented_images` affiche un échantillon d'images augmentées pour visualiser les transformations appliquées.


```python
from tensorflow.keras import layers

# Appliquer le recadrage et le retournement horizontal
data_augmentation = tf.keras.Sequential([
    layers.RandomCrop(32, 32),
    layers.RandomFlip('horizontal'),
])

# Afficher les images originales
def plot_images(images, labels, class_names, num_images=10):
    plt.figure(figsize=(12, 12))
    for i in range(num_images):
        plt.subplot(1, num_images, i + 1)
        plt.imshow(images[i])
        plt.title(class_names[labels[i]])
        plt.axis('off')
    plt.show()

class_names = sorted(os.listdir(TRAIN_FOLDER))  # Récupération des noms de classes
plot_images(train_images, train_labels, class_names)

train_images = data_augmentation(train_images)
test_images = data_augmentation(test_images)
val_images = data_augmentation(val_images)

# Afficher les formes après augmentation
print(f"Forme des images d'entraînement après augmentation: {train_images.shape}")
print(f"Forme des images de test après augmentation: {test_images.shape}")

# Afficher les images augmentées
def plot_augmented_images(images, data_augmentation, num_images=10):
    augmented_images = [data_augmentation(tf.expand_dims(image, 0))[0] for image in images[:num_images]]
    
    plt.figure(figsize=(12, 12))
    for i in range(num_images):
        plt.subplot(1, num_images, i + 1)
        plt.imshow(augmented_images[i])
        plt.axis('off')
    plt.show()

plot_augmented_images(train_images, data_augmentation)
```

### Définition et Compilation du Modèle de Réseau de Neurones

Le code suivant construit et compile un modèle de réseau de neurones convolutifs pour la classification d'images. Le modèle est conçu avec plusieurs couches convolutives pour augmenter sa profondeur et améliorer sa capacité à extraire des caractéristiques des images.

- **Définition du modèle** :
  - `Conv2D` : Couches de convolution pour extraire des caractéristiques à partir des images.
  - `MaxPooling2D` : Couches de sous-échantillonnage pour réduire la dimensionnalité des caractéristiques extraites.
  - `Flatten` : Transformation des données 2D en un vecteur 1D.
  - `Dense` : Couches entièrement connectées pour la classification finale.

- **Compilation du modèle** :
  - **Optimiseur** : `Adam` avec un taux d'apprentissage de 0.001, ajusté pour optimiser le modèle.
  - **Fonction de perte** : `sparse_categorical_crossentropy`, utilisée pour la classification multi-classe avec des étiquettes entières.
  - **Métriques** : `accuracy`, pour évaluer la performance du modèle pendant l'entraînement et la validation.


```python
from tensorflow.keras import layers, models

# Augmenter la profondeur du modèle
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),  # Ajuster le taux d'apprentissage
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

### Entraînement du Modèle

Le code suivant entraîne le modèle de réseau de neurones sur les images d'entraînement et évalue sa performance sur les données de validation. Il est configuré pour effectuer 20 époques d'entraînement.

- **Entraînement** :
  - `train_images` et `train_labels` : Données d'entraînement utilisées pour ajuster les poids du modèle.
  - `validation_data` : Données de validation utilisées pour évaluer la performance du modèle après chaque époque.
  - `epochs` : Nombre total d'époques pour l'entraînement, ici fixé à 20.

- **Sortie** :
  - `history` : Objet retourné par la méthode `fit`, contenant les informations sur la perte et la précision du modèle au cours de l'entraînement et de la validation.

```python
history = model.fit(train_images, train_labels, epochs=20,
                    validation_data=(val_images, val_labels))
```

### Visualisation des Performances du Modèle

Le code suivant génère des graphiques pour évaluer la performance du modèle au cours de l'entraînement. Il montre l'évolution de la perte et de la précision pendant l'entraînement et la validation, ainsi qu'une relation entre précision et perte.

- **Graphiques des pertes et précisions** :
  - **Pertes** : Affiche la perte d'entraînement et de validation en fonction des époques.
  - **Précision** : Affiche la précision d'entraînement et de validation en fonction des époques.

- **Graphique Précision vs Perte** :

```python
# Graphiques des pertes et précisions
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Perte entraînement')
plt.plot(history.history['val_loss'], label='Perte validation')
plt.title('Pertes pendant l\'entraînement')
plt.xlabel('Époques')
plt.ylabel('Perte')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Précision entraînement')
plt.plot(history.history['val_accuracy'], label='Précision validation')
plt.title('Précision pendant l\'entraînement')
plt.xlabel('Époques')
plt.ylabel('Précision')
plt.legend()

plt.show()

# Graphique Précision vs Perte
plt.figure()
plt.plot(history.history['accuracy'], history.history['loss'], 'o-')
plt.title('Précision vs Perte')
plt.xlabel('Précision')
plt.ylabel('Perte')
plt.show()
```

### Prédictions et Visualisation des Résultats

Le code suivant utilise le modèle pour faire des prédictions sur les images de test, puis affiche ces images avec leurs étiquettes véritables et prédites. 

- **Prédictions** :
  - Le modèle prédit les étiquettes des images de test.
  - Les étiquettes prédites sont obtenues en prenant l'indice de la valeur maximale dans les résultats de prédiction.

- **Affichage des images avec les prédictions** :
  - **Fonction `plot_predictions`** : Affiche les images de test avec les étiquettes véritables et les étiquettes prédites par le modèle.
    - Les titres des images montrent les étiquettes véritables et prédites.
    - Le titre est affiché avec un fond opaque pour une meilleure lisibilité.

```python
predictions = model.predict(test_images)
predictions_labels = np.argmax(predictions, axis=1)

# Afficher les images avec les prédictions
def plot_predictions(images, labels, predictions, class_names, num_images=10):
    plt.figure(figsize=(15, 15))
    for i in range(num_images):
        plt.subplot(1, num_images, i + 1)
        plt.imshow(images[i])
        
        predicted_label = np.argmax(predictions[i])
        true_label = labels[i]
        
        title_text = f'True: {class_names[true_label]}\nPred: {class_names[predicted_label]}'
        
        plt.title(title_text, fontsize=10, pad=5)
        plt.axis('off')
        
        # Ajouter un fond opaque pour le titre
        plt.gca().set_title(title_text, fontsize=10, pad=5, backgroundcolor='white', color='black')
        
    plt.tight_layout()
    plt.show()

plot_predictions(test_images, test_labels, predictions, class_names)
```

### Rapport de Classification

Le code suivant génère un rapport de classification qui fournit des métriques détaillées sur la performance du modèle, telles que la précision, le rappel, et la mesure F1 pour chaque classe. Ce rapport est utile pour évaluer la qualité des prédictions du modèle sur les données de test.

### Analyse de la Courbe ROC et de l'AUC

Le code suivant calcule et trace les courbes ROC (Receiver Operating Characteristic) et les valeurs AUC (Area Under the Curve) pour chaque classe. Ces courbes aident à évaluer la performance du modèle en termes de capacité à séparer les classes positives des classes négatives.

- **Préparation des données** :
  - `y_true` : Étiquettes vraies des images de test.
  - `y_scores` : Scores de prédiction pour chaque classe.

- **Calcul de ROC et AUC** :
  - Pour chaque classe, la courbe ROC est calculée à l'aide des taux de faux positifs (FPR) et des taux de vrais positifs (TPR).
  - L'AUC est calculée comme l'aire sous la courbe ROC pour chaque classe.

- **Tracé des courbes ROC** :
  - Chaque courbe ROC est tracée avec son AUC respectif.
  - Une ligne diagonale (`k--`) est ajoutée pour représenter une performance aléatoire.
  - Les courbes sont limitées entre 0 et 1 pour les axes.

```python
# Préparer les données
y_true = test_labels
y_scores = predictions

# Calcul de la ROC et AUC pour chaque classe
n_classes = y_scores.shape[1]
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    # Calculer ROC uniquement si les valeurs sont suffisantes pour la classe
    if np.sum(y_true == i) > 0 and np.sum(y_scores[:, i] > 0) > 0:
        fpr[i], tpr[i], _ = roc_curve(y_true, y_scores[:, i], pos_label=i)
        roc_auc[i] = auc(fpr[i], tpr[i])
    else:
        fpr[i] = np.array([])
        tpr[i] = np.array([])
        roc_auc[i] = np.nan

# Tracer les courbes ROC
plt.figure(figsize=(12, 8))
for i in range(n_classes):
    if len(fpr[i]) > 0 and len(tpr[i]) > 0:
        plt.plot(fpr[i], tpr[i], lw=2, label=f'Classe {i} (AUC = {roc_auc[i]:.2f})')
    else:
        print(f"Skipping plot for class {i} due to insufficient data.")

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

### Analyse du Modèle et Affichage des Activations des Couches

Le code suivant fournit une vue d'ensemble du modèle en affichant sa structure, ses formes d'entrée et de sortie pour chaque couche, et les activations des premières couches. Ces étapes permettent d'examiner comment les données se propagent à travers le modèle et de visualiser les activations des couches pour comprendre ce que le modèle apprend à différents niveaux.

- **Résumé du Modèle** :
  - `model.summary()` affiche un résumé détaillé du modèle, incluant les couches, les paramètres, et les formes des entrées et sorties.

- **Vérification des Formes d'Entrée et de Sortie des Couches** :
  - Pour chaque couche, les formes d'entrée et de sortie sont affichées pour vérifier que les dimensions sont correctes.

- **Préparation des Activations** :
  - Les sorties des premières couches sont extraites pour être visualisées.
  - Un modèle `activation_model` est créé pour obtenir les activations intermédiaires à partir des premières images de test.

- **Affichage des Activations des Couches** :
  - La fonction `display_layer_activations` visualise les activations des premières couches. Chaque activation est normalisée et affichée en tant que grille d'images pour chaque couche.

```python
model.summary()

# Assurez-vous d'avoir toutes les couches correctement initialisées
for layer in model.layers:
    print(f"{layer.name}: {layer.get_input_shape_at(0)}, {layer.get_output_shape_at(0)}")

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models

_ = model.predict(test_images[:1])

# Afficher les formes d'entrée et de sortie des couches
for layer in model.layers:
    # Utilisation des attributs `input_shape` et `output_shape` directement
    input_shape = layer.input_shape
    output_shape = layer.output_shape
    print(f"{layer.name}: input_shape={input_shape}, output_shape={output_shape}")

# Obtenir les sorties des premières couches du modèle
layer_outputs = [layer.output for layer in model.layers[:6] if layer.output is not None]

# Créer un modèle d'activations en utilisant des sorties intermédiaires
activation_model = models.Model(inputs=model.input, outputs=layer_outputs)

# Obtenir les activations des premières images de test
activations = activation_model.predict(test_images[:1])

# Récupérer les noms des couches
layer_names = [layer.name for layer in model.layers[:6] if layer.output is not None]

# Fonction pour afficher les activations des couches
def display_layer_activations(activations, layer_names):
    images_per_row = 16
    for layer_name, layer_activation in zip(layer_names, activations):
        n_features = layer_activation.shape[-1]
        size = layer_activation.shape[1]
        n_cols = n_features // images_per_row
        display_grid = np.zeros((size * n_cols, images_per_row * size))
        
        for col in range(n_cols):
            for row in range(images_per_row):
                channel_image = layer_activation[0, :, :, col * images_per_row + row]
                channel_image -= channel_image.mean()
                channel_image /= channel_image.std()
                channel_image *= 64
                channel_image += 128
                channel_image = np.clip(channel_image, 0, 255).astype('uint8')
                display_grid[col * size: (col + 1) * size,
                             row * size: (row + 1) * size] = channel_image
        
        scale = 1. / size
        plt.figure(figsize=(scale * display_grid.shape[1],
                            scale * display_grid.shape[0]))
        plt.title(layer_name)
        plt.grid(False)
        plt.imshow(display_grid, aspect='auto', cmap='viridis')
        plt.show()

# Afficher les activations des couches
display_layer_activations(activations, layer_names)
```
