```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, roc_curve, auc, confusion_matrix, classification_report
from tensorflow.keras.callbacks import EarlyStopping

print("imported")
```

```python
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
```

```python
IMG_HEIGHT, IMG_WIDTH = 128, 128
BATCH_SIZE = 32
EPOCHS = 10 
FOLDS = 10   
DATA_DIR = '../../../chest_xray/train'

print("parametres defined")
```

```python
# Chargement des données
def load_data(data_dir):
    print("chargement des données...")

    datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

    print("générateur de données d'entraînement...")
    train_data = datagen.flow_from_directory(
        data_dir,
        target_size=(IMG_HEIGHT, IMG_WIDTH),
        color_mode='grayscale',
        batch_size=BATCH_SIZE,
        class_mode='binary',
        subset='training'
    )
    print("données d'entraînement chargées")

    print("générateur de données de validation...")
    val_data = datagen.flow_from_directory(
        data_dir,
        target_size=(IMG_HEIGHT, IMG_WIDTH),
        color_mode='grayscale',
        batch_size=BATCH_SIZE,
        class_mode='binary',
        subset='validation'
    )
    print("données de validation chargées")

    return train_data, val_data

train_data, val_data = load_data(DATA_DIR)
print("Chargement des données terminé.")
```

```python
# Définition du modèle CNN modifié
def create_model():
    print("debut de la création du modèle...")

    model = models.Sequential()

    print("ajout de la couche d'entrée...")
    model.add(layers.Input(shape=(IMG_HEIGHT, IMG_WIDTH, 1)))  # Grayscale

    print("ajout des couches de convolution et de pooling...")
    for filters in [32, 64, 128, 256, 512]:
        model.add(layers.Conv2D(filters, (3, 3), activation='relu'))
        model.add(layers.MaxPooling2D((2, 2)))
        print(f"Couche de convolution ajoutée avec {filters} filtres.")

    print("ajout de la couche Flatten...")
    model.add(layers.Flatten())

    print("ajout des couches denses avec Dropout...")
    for units in [512, 256, 128]:
        model.add(layers.Dense(units, activation='relu'))
        model.add(layers.Dropout(0.5))
        print(f"Couche dense ajoutée avec {units} unités.")

    print("ajout de la couche de sortie...")
    model.add(layers.Dense(1, activation='sigmoid'))

    print("compilation du modèle...")
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', tf.keras.metrics.AUC()])

    print("modèle créé et compilé.")
    return model
```

```python
# Validation croisée K-Fold
def k_fold_cross_validation(train_data, val_data):
    kf = KFold(n_splits=FOLDS, shuffle=True)
    accuracy_list = []
    auc_list = []
    all_histories = []

    # boucle à travers les plis
    for fold, (train_index, val_index) in enumerate(kf.split(train_data)):
        print(f"Fold {fold+1}/{FOLDS}")

        # création du modèle
        model = create_model()
        print(f"Modèle créé pour le pli {fold+1}")

        # configuration de l'EarlyStopping
        early_stopping = EarlyStopping(monitor='val_loss', patience=3)
        print(f"EarlyStopping configuré pour le pli {fold+1}")

        # entraînement du modèle
        history = model.fit(train_data, validation_data=val_data, epochs=EPOCHS, callbacks=[early_stopping], verbose=0)
        all_histories.append(history.history)
        print(f"Modèle entraîné pour le pli {fold+1}")

        # évaluation du modèle
        results = model.evaluate(val_data)
        val_loss = results[0]
        val_accuracy = results[1]
        val_auc = results[2]
        accuracy_list.append(val_accuracy)
        auc_list.append(val_auc)
        print(f"Validation accuracy for fold {fold+1}: {val_accuracy:.4f}")
        print(f"AUC for fold {fold+1}: {val_auc:.4f}")

    return np.mean(accuracy_list), np.mean(auc_list), all_histories


# entraînement et validation croisée
accuracy, roc_auc, histories = k_fold_cross_validation(train_data, val_data)
print(f'Mean Accuracy: {accuracy:.4f}, Mean AUC: {roc_auc:.4f}')
```

```python
# Visualisation des résultats avec titres et séparations
def plot_results(histories):
    # Graphiques de la précision
    plt.figure(figsize=(12, 6))
    for i, history in enumerate(histories):
        plt.plot(history['accuracy'], label=f'Fold {i+1} Training Accuracy')
        plt.plot(history['val_accuracy'], label=f'Fold {i+1} Validation Accuracy')
    
    plt.title('Model Training and Validation Accuracy per Fold')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(loc='upper left')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    
    # Graphiques de la perte
    plt.figure(figsize=(12, 6))
    for i, history in enumerate(histories):
        plt.plot(history['loss'], label=f'Fold {i+1} Training Loss')
        plt.plot(history['val_loss'], label=f'Fold {i+1} Validation Loss')
    
    plt.title('Model Training and Validation Loss per Fold')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(loc='upper right')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Appel de la fonction
plot_results(histories)
```

```python
model = create_model()
model.fit(train_data, validation_data=val_data, epochs=EPOCHS)
model.save('cnn.h5')
```

```python
# Évaluation finale sur le jeu de test
test_data_dir = '../../../chest_xray/test'
test_data = ImageDataGenerator(rescale=1./255).flow_from_directory(
    test_data_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    color_mode='grayscale',
    batch_size=BATCH_SIZE,
    class_mode='binary'
)
test_loss, test_accuracy = model.evaluate(test_data)
print(f"Test Accuracy: {test_accuracy:.4f}")
```

```python
# Courbe ROC et matrice de confusion
y_true = test_data.classes
y_scores = model.predict(test_data).ravel()
fpr, tpr, _ = roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.2f}')
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid()
plt.show()

cm = confusion_matrix(y_true, (y_scores > 0.5).astype(int))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Rapport de classification
print(classification_report(y_true, (y_scores > 0.5).astype(int)))
```
