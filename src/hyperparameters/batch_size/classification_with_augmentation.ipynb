```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import seaborn as sns
```

```python
if tf.config.list_physical_devices('GPU'):
    print("GPU détecté.")
else:
    print("Aucun GPU détecté, l'entraînement se fera sur le CPU.")
```

```python
# Paramètres
img_size = 150
data_dir1 = '../../../chest_xray/train/PNEUMONIA'
data_dir2 = '../../../chest_xray/train/NORMAL'
batch_size = 128  # Taille du batch augmentée
epochs = 20
padding_size = 20 
```

```python
# Fonction pour charger les données avec débogage
def load_data(data_dir1, data_dir2, limit=100):
    print("Chargement des données...")
    data = []
    labels = ['PNEUMONIA', 'NORMAL']
    
    # Vérification et affichage des fichiers dans chaque répertoire
    for dir in [data_dir1, data_dir2]:
        path = os.path.join(dir)
        if not os.path.exists(path):
            print(f"Le répertoire {path} n'existe pas.")
            continue
        
        print(f"Fichiers dans {path}: {os.listdir(path)[:5]} ...")  # Affiche les 5 premiers fichiers
        
        label = os.path.basename(dir)
        class_num = labels.index(label)
        for i, img in enumerate(os.listdir(path)):
            if i >= limit:
                break
            try:
                img_path = os.path.join(path, img)
                img_arr = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
                
                if img_arr is None:
                    print(f"Impossible de lire {img_path}. Ignoré.")
                    continue
                
                resized_arr = cv2.resize(img_arr, (img_size, img_size))
                data.append([resized_arr, class_num])
            except Exception as e:
                print(f"Erreur lors du traitement de l'image {img}: {e}")
    
    print(f"Chargement des données terminé. {len(data)} images chargées.")
    return np.array(data, dtype=object)

# Charger les données
train_data = load_data(data_dir1, data_dir2, limit=100)
```

```python
# Vérifier si les données sont bien chargées
if len(train_data) == 0:
    raise ValueError("Aucune donnée n'a été chargée. Vérifiez les chemins d'accès et le contenu des répertoires.")
else:
    print("Les données ont été chargées avec succès.")
```

```python
# Séparer les caractéristiques et les labels
def separate_features_labels(data):
    features = []
    labels = []
    for img, label in data:
        features.append(img)
        labels.append(label)
    return np.array(features), np.array(labels)

# Prétraiter les images
def preprocess_image(image):
    if len(image.shape) == 2:
        image = np.expand_dims(image, axis=-1)
    image = tf.image.resize(image, [img_size, img_size])
    image = image / 255.0
    return image

# Prétraiter les données
def preprocess_data(data):
    processed_data = []
    for img, label in data:
        img = preprocess_image(img)
        processed_data.append([img, label])
    return np.array(processed_data, dtype=object)

# Prétraitement
train_data = preprocess_data(train_data)
x_train, y_train = separate_features_labels(train_data)

# Reshape des données pour être compatibles avec les CNN
x_train = x_train.reshape(-1, img_size, img_size, 1)
x_train = np.concatenate([x_train, x_train, x_train], axis=-1)  # Convertir en 3 canaux

print(f"x_train shape: {x_train.shape}")
print(f"y_train shape: {y_train.shape}")

# Diviser les données en ensembles d'entraînement et de validation
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
```

```python
# Vérifier que les ensembles d'entraînement et de validation ne sont pas vides
if len(x_train) == 0 or len(x_val) == 0:
    raise ValueError("L'ensemble d'entraînement ou de validation est vide. Vérifiez les données et la taille du split.")
else:
    print("Les ensembles d'entraînement et de validation ont été chargés avec succès.")
```

```python
# Initialisation de l'ImageDataGenerator
datagen = ImageDataGenerator(
    zoom_range=0.1,
    shear_range=0.1
)
datagen.fit(x_train)
```

```python
# Définir le modèle CNN
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(img_size, img_size, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), 
              loss='binary_crossentropy', 
              metrics=['accuracy'])
print("model compiled")

model.build(input_shape=(None, img_size, img_size, 3))
print("model built")
```

```python
# Callbacks pour l'entraînement
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=0.0001)
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
history = model.fit(datagen.flow(x_train, y_train, batch_size=batch_size),
                    epochs=epochs,
                    validation_data=(x_val, y_val),
                    callbacks=[reduce_lr, early_stop])
```

```python
# Visualiser les courbes d'apprentissage
def plot_learning_curves(history):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()

plot_learning_curves(history)
```

```python
scores = model.evaluate(x_val, y_val, verbose=0)
print(f"Validation accuracy: {scores[1]*100:.2f}%")
print(f"Validation loss: {scores[0]:.4f}")
```

```python
# Matrice de confusion et rapport de classification
y_pred = (model.predict(x_val) > 0.5).astype("int32")
cm = confusion_matrix(y_val, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['PNEUMONIA', 'NORMAL'], yticklabels=['PNEUMONIA', 'NORMAL'])
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.show()

cr = classification_report(y_val, y_pred, target_names=['PNEUMONIA', 'NORMAL'])
print("Classification Report:\n", cr)
```

```python
# Courbe ROC
fpr, tpr, _ = roc_curve(y_val, y_pred)
roc_auc = auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
# Distribution des classes
def plot_class_distribution(y_data, title):
    plt.figure(figsize=(8, 6))
    sns.histplot(y_data, discrete=True, kde=False, bins=2)
    plt.xlabel('Classe')
    plt.ylabel('Nombre d\'échantillons')
    plt.title(title)
    plt.xticks([0, 1], ['PNEUMONIA', 'NORMAL'])
    plt.show()

plot_class_distribution(y_train, 'Distribution des classes dans l\'ensemble d\'entraînement')
plot_class_distribution(y_val, 'Distribution des classes dans l\'ensemble de validation')
```

```python
import pandas as pd

# Fonction pour tracer l'histogramme des prédictions
def plot_prediction_histogram(y_true, y_pred_probs):
    plt.figure(figsize=(10, 6))
    plt.hist(y_pred_probs[y_true == 0], bins=30, alpha=0.5, label='Normal', color='blue')
    plt.hist(y_pred_probs[y_true == 1], bins=30, alpha=0.5, label='Pneumonia', color='red')
    plt.xlabel('Probability')
    plt.ylabel('Frequency')
    plt.title('Histogram of Predicted Probabilities')
    plt.legend(loc='upper right')
    plt.show()

plot_prediction_histogram(y_val, y_pred_probs)
```

```python
# Fonction pour afficher le tableau des prédictions
def print_predictions_table(y_true, y_pred_probs):
    # Convertir les prédictions en classes
    y_pred = (y_pred_probs > 0.5).astype(int)
    # Créer un DataFrame pour afficher les résultats
    df = pd.DataFrame({'True Label': y_true, 'Predicted Label': y_pred.flatten(), 'Predicted Probability': y_pred_probs.flatten()})
    print(df.head(20))  # Afficher les 20 premières lignes

# Tracer l'histogramme des prédictions
y_pred_probs = model.predict(x_val)

# Afficher le tableau des prédictions
print_predictions_table(y_val, y_pred_probs)
```
