```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import seaborn as sns
from keras_tuner.tuners import Hyperband
from keras_tuner.engine.hyperparameters import HyperParameters
```

```python
if tf.config.list_physical_devices('GPU'):
    print("GPU détecté.")
else:
    print("Aucun GPU détecté, l'entraînement se fera sur le CPU.")
```

```python
# Paramètres
img_size = 32
batch_size = 32
epochs = 20
data_dir1 = '../../../chest_xray/train/PNEUMONIA'
data_dir2 = '../../../chest_xray/train/NORMAL'
```

```python
def load_data(data_dir1, data_dir2, limit=100):
    print("Chargement des données...")
    data = []
    labels = ['PNEUMONIA', 'NORMAL']
    
    for dir in [data_dir1, data_dir2]:
        path = os.path.join(dir)
        if not os.path.exists(path):
            print(f"Le répertoire {path} n'existe pas.")
            continue
        
        print(f"Fichiers dans {path}: {os.listdir(path)[:5]} ...")
        
        label = os.path.basename(dir)
        class_num = labels.index(label)
        for i, img in enumerate(os.listdir(path)):
            if i >= limit:
                break
            try:
                img_path = os.path.join(path, img)
                img_arr = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
                
                if img_arr is None:
                    print(f"Impossible de lire {img_path}. Ignoré.")
                    continue
                
                resized_arr = cv2.resize(img_arr, (img_size, img_size))
                data.append([resized_arr, class_num])
            except Exception as e:
                print(f"Erreur lors du traitement de l'image {img}: {e}")
    
    print(f"Chargement des données terminé. {len(data)} images chargées.")
    return np.array(data, dtype=object)

# Charger les données
train_data = load_data(data_dir1, data_dir2, limit=100)
if len(train_data) == 0:
    raise ValueError("Aucune donnée n'a été chargée. Vérifiez les chemins d'accès et le contenu des répertoires.")
else:
    print("Les données ont été chargées avec succès.")
```

```python
# Séparer les caractéristiques et les labels
def separate_features_labels(data):
    features = []
    labels = []
    for img, label in data:
        features.append(img)
        labels.append(label)
    return np.array(features), np.array(labels)

# Prétraiter les images
def preprocess_image(image):
    if len(image.shape) == 2:
        image = np.expand_dims(image, axis=-1)
    image = tf.image.resize(image, [img_size, img_size])
    image = image / 255.0
    return image

# Prétraiter les données
def preprocess_data(data):
    processed_data = []
    for img, label in data:
        img = preprocess_image(img)
        processed_data.append([img, label])
    return np.array(processed_data, dtype=object)

# Prétraitement
train_data = preprocess_data(train_data)
x_train, y_train = separate_features_labels(train_data)

# Reshape des données pour être compatibles avec les CNN
x_train = x_train.reshape(-1, img_size, img_size, 1)
x_train = np.concatenate([x_train, x_train, x_train], axis=-1)  # Convertir en 3 canaux

# Diviser les données en ensembles d'entraînement et de validation
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
if len(x_train) == 0 or len(x_val) == 0:
    raise ValueError("L'ensemble d'entraînement ou de validation est vide. Vérifiez les données et la taille du split.")
else:
    print("Les ensembles d'entraînement et de validation ont été chargés avec succès.")
```

```python
print("Ajout de la dimension des canaux à x_train...")
x_train = x_train.reshape(-1, img_size, img_size, 1) 
print(f"Forme de x_train après ajout de la dimension des canaux : {x_train.shape}")

print("Conversion de x_train en 3 canaux...")
x_train = np.concatenate([x_train, x_train, x_train], axis=-1) 
print(f"Forme de x_train après conversion en 3 canaux : {x_train.shape}")

print("Création de l'ImageDataGenerator...")
datagen = ImageDataGenerator(
    rotation_range=10,
    zoom_range=0.1,
    shear_range=0.1,
    horizontal_flip=True,
    fill_mode="nearest"
)

print("Ajustement des données d'entraînement avec ImageDataGenerator...")
datagen.fit(x_train)
print("Ajustement des données d'entraînement terminé.")
```

```python
# Vérifiez les dimensions
print(f"Dimensions de x_train : {x_train.shape}")
print(f"Dimensions de y_train : {y_train.shape}")
print(f"Dimensions de x_val : {x_val.shape}")
print(f"Dimensions de y_val : {y_val.shape}")
```

```python
# Initialisation de l'ImageDataGenerator
datagen = ImageDataGenerator(
    rotation_range=10,
    zoom_range=0.1,
    shear_range=0.1,
    horizontal_flip=True,
    fill_mode="nearest"
)
datagen.fit(x_train)
```

```python
def create_model(hp):
    model = Sequential([
        Conv2D(hp.Int('filters_1', min_value=16, max_value=64, step=16), (3, 3), activation='relu', input_shape=(img_size, img_size, 3)),
        MaxPooling2D((2, 2)),
        Conv2D(hp.Int('filters_2', min_value=32, max_value=128, step=32), (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Conv2D(hp.Int('filters_3', min_value=64, max_value=256, step=64), (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(hp.Int('units', min_value=64, max_value=256, step=64), activation='relu'),
        Dropout(hp.Float('dropout', min_value=0.2, max_value=0.5, step=0.1)),
        Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer=tf.keras.optimizers.Adam(hp.Float('learning_rate', min_value=1e-4, max_value=1e-2, sampling='LOG')),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model
```

```python
def perform_grid_search(x_train, y_train):
    tuner = Hyperband(
        create_model,
        objective='val_accuracy',
        max_epochs=20,
        factor=3,
        directory='my_dir',
        project_name='helloworld'
    )

    tuner.search(x_train, y_train, epochs=20, validation_data=(x_val, y_val))
    best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
    print(f"Meilleurs hyperparamètres : {best_hps.values}")
    return tuner
```

```python
# Effectuer la recherche d'hyperparamètres
tuner = perform_grid_search(x_train, y_train)
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
```

```python
# Construire et entraîner le modèle avec les meilleurs hyperparamètres
model = create_model(best_hps)
history = model.fit(datagen.flow(x_train, y_train, batch_size=batch_size),
                    epochs=epochs,  # Utilisez la valeur fixe définie précédemment
                    validation_data=(x_val, y_val),
                    callbacks=[
                        ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=0.0001),
                        EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
                    ])
```

```python
# Sauvegarder le modèle
model.save('best_model.h5')
print("Modèle sauvegardé sous 'best_model.h5'")
```

```python
# Évaluation du modèle
scores = model.evaluate(x_val, y_val, verbose=0)
print(f"Validation accuracy: {scores[1]*100:.2f}%")
print(f"Validation loss: {scores[0]:.4f}")
```

```python
# Matrice de confusion et rapport de classification
y_pred = (model.predict(x_val) > 0.5).astype("int32")
cm = confusion_matrix(y_val, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['PNEUMONIA', 'NORMAL'], yticklabels=['PNEUMONIA', 'NORMAL'])
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.show()

cr = classification_report(y_val, y_pred, target_names=['PNEUMONIA', 'NORMAL'])
print("Classification Report:\n", cr)
```

```python
# Courbe ROC
fpr, tpr, _ = roc_curve(y_val, model.predict(x_val))
roc_auc = auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()
```

```python
def plot_class_distribution(y_train, y_val):
    plt.figure(figsize=(12, 6))
    
    # Distribution dans l'ensemble d'entraînement
    plt.subplot(1, 2, 1)
    sns.histplot(y_train, bins=2, kde=False, palette="viridis", discrete=True)
    plt.title('Distribution des classes (Entraînement)')
    plt.xlabel('Classe')
    plt.ylabel('Nombre d\'exemples')
    
    # Distribution dans l'ensemble de validation
    plt.subplot(1, 2, 2)
    sns.histplot(y_val, bins=2, kde=False, palette="viridis", discrete=True)
    plt.title('Distribution des classes (Validation)')
    plt.xlabel('Classe')
    plt.ylabel('Nombre d\'exemples')
    
    plt.tight_layout()
    plt.show()

# Exemple d'utilisation
plot_class_distribution(y_train, y_val)
```

```python
def plot_prediction_histogram(y_true, y_pred_prob):
    plt.figure(figsize=(12, 6))
    
    # Convertir les probabilités en étiquettes binaires
    y_pred_labels = (y_pred_prob > 0.5).astype(int)
    
    # Histogramme des probabilités prédites
    sns.histplot(y_pred_prob, bins=30, kde=True, color='blue', label='Probabilités prédites')
    plt.title('Histogramme des probabilités prédites')
    plt.xlabel('Probabilité')
    plt.ylabel('Nombre d\'exemples')
    plt.legend()
    plt.show()

# Exemple d'utilisation
y_pred_prob = model.predict(x_val)  # Probabilités prédites pour l'ensemble de validation
plot_prediction_histogram(y_val, y_pred_prob)
```

```python
import pandas as pd

def print_predictions_table(y_true, y_pred_prob):
    y_pred_labels = (y_pred_prob > 0.5).astype(int)
    
    # Créer un DataFrame avec les vraies étiquettes, les prédictions, et les probabilités
    df = pd.DataFrame({
        'True Label': y_true,
        'Predicted Label': y_pred_labels.flatten(),
        'Predicted Probability': y_pred_prob.flatten()
    })
    
    # Afficher les premières lignes du DataFrame
    print(df.head())

# Exemple d'utilisation
print_predictions_table(y_val, y_pred_prob)
```
